```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By  
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Path to your Edge WebDriver executable
PATH = "/home/asif/Downloads/edgedriver_linux64/msedgedriver"

# Create a Service object for Edge WebDriver
service = Service(executable_path=PATH)

# Initialize the Edge WebDriver with the Service object
driver = webdriver.Edge(service=service)

# Open the webpage
driver.get("https://orteil.dashnet.org/cookieclicker/")

# Wait for the language selection element and click it
language = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.ID, "langSelect-EN"))
)
language.click()

# Wait for the page to load fully
WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.ID, "bigCookie"))
)
# driver.implicitly_wait(10)


# Find the cookie and the cookie count element
cookie = driver.find_element(By.ID, "bigCookie")
cookie_count = driver.find_element(By.ID, "cookies")

# Perform the cookie clicking action
# for i in range(5000):
#     cookie.click()
#     print(cookie_count.text)
#     time.sleep(0.1)  # Small delay to avoid overwhelming the page
    
while True:
    cookie.click()
    cookie_count = driver.find_element(By.ID, "cookies").text.split(" ")[0]
    print(cookie_count)
    cookie_count = int(cookie_count)
    product_price_prefix = "productPrice"
    for i in range(4):
        product_price= driver.find_element(By.ID, product_price_prefix + str(i)).text.split(" ")[0]
        if not product_price.isdigit():
            continue
        product_price = int(product_price)
        if cookie_count >= product_price:
            product = driver.find_element(By.ID, "product" + str(i))
            product.click()
            break
        

# Optional: Close the browser after the task
# driver.quit()
```
